### Proposition — The Riemann Curvature Tensor Transforms Tensorially

Let $x^\mu$ and $x'^\mu$ be two coordinate systems. Define

$$
A^\alpha_{\ \rho}
=
\frac{\partial x^\alpha}{\partial x'^\rho},
\qquad
B^\tau_{\ \lambda}
=
\frac{\partial x'^\tau}{\partial x^\lambda}.
$$

Then

$$
B^\tau_{\ \lambda}A^\lambda_{\ \rho}
=
\delta^\tau_{\rho},
\qquad
A^\alpha_{\ \rho}B^\rho_{\ \beta}
=
\delta^\alpha_{\beta}.
$$

The Christoffel symbols transform as

$$
\Gamma'^\tau_{\rho\sigma}
=
B^\tau_{\ \lambda}
A^\mu_{\ \rho}
A^\nu_{\ \sigma}
\Gamma^\lambda_{\mu\nu}
+
B^\tau_{\ \lambda}
\frac{\partial^2 x^\lambda}{\partial x'^\rho\partial x'^\sigma}.
$$

Define the Riemann curvature tensor by

$$
R^\tau_{\ \rho\mu\nu}
=
\frac{\partial \Gamma^\tau_{\rho\nu}}{\partial x^\mu}
-
\frac{\partial \Gamma^\tau_{\rho\mu}}{\partial x^\nu}
+
\Gamma^\tau_{\lambda\mu}\Gamma^\lambda_{\rho\nu}
-
\Gamma^\tau_{\lambda\nu}\Gamma^\lambda_{\rho\mu}.
$$

Then

$$
R'^\tau_{\ \rho\mu\nu}
=
B^\tau_{\ \alpha}
A^\beta_{\ \rho}
A^\gamma_{\ \mu}
A^\delta_{\ \nu}
R^\alpha_{\ \beta\gamma\delta}.
$$

Equivalently,

$$
R'^\tau_{\ \rho\mu\nu}
=
\frac{\partial x'^\tau}{\partial x^\alpha}
\frac{\partial x^\beta}{\partial x'^\rho}
\frac{\partial x^\gamma}{\partial x'^\mu}
\frac{\partial x^\delta}{\partial x'^\nu}
R^\alpha_{\ \beta\gamma\delta}.
$$

<!-- proof:start -->
### Proof

We start from the definition of the Riemann tensor in the primed coordinates:

$$
R'^\tau_{\ \rho\mu\nu}
=
\frac{\partial \Gamma'^\tau_{\rho\nu}}{\partial x'^\mu}
-
\frac{\partial \Gamma'^\tau_{\rho\mu}}{\partial x'^\nu}
+
\Gamma'^\tau_{\lambda\mu}\Gamma'^\lambda_{\rho\nu}
-
\Gamma'^\tau_{\lambda\nu}\Gamma'^\lambda_{\rho\mu}.
$$

For compactness, write

$$
\partial'_\mu
=
\frac{\partial}{\partial x'^\mu},
\qquad
\partial_\gamma
=
\frac{\partial}{\partial x^\gamma}.
$$

By the chain rule,

$$
\partial'_\mu
=
A^\gamma_{\ \mu}\partial_\gamma.
$$

The Christoffel transformation law is

$$
\Gamma'^\tau_{\rho\nu}
=
B^\tau_{\ \alpha}
A^\beta_{\ \rho}
A^\delta_{\ \nu}
\Gamma^\alpha_{\beta\delta}
+
B^\tau_{\ \alpha}
\partial'_\rho A^\alpha_{\ \nu}.
$$

This follows because

$$
A^\alpha_{\ \nu}
=
\frac{\partial x^\alpha}{\partial x'^\nu},
$$

so that

$$
\partial'_\rho A^\alpha_{\ \nu}
=
\frac{\partial^2 x^\alpha}{\partial x'^\rho\partial x'^\nu}.
$$

Now compute the first derivative term:

$$
\partial'_\mu \Gamma'^\tau_{\rho\nu}
=
\partial'_\mu
\left(
B^\tau_{\ \alpha}
A^\beta_{\ \rho}
A^\delta_{\ \nu}
\Gamma^\alpha_{\beta\delta}
\right)
+
\partial'_\mu
\left(
B^\tau_{\ \alpha}
\partial'_\rho A^\alpha_{\ \nu}
\right).
$$

Using the product rule on the first term,

$$
\partial'_\mu
\left(
B^\tau_{\ \alpha}
A^\beta_{\ \rho}
A^\delta_{\ \nu}
\Gamma^\alpha_{\beta\delta}
\right)
=
(\partial'_\mu B^\tau_{\ \alpha})
A^\beta_{\ \rho}
A^\delta_{\ \nu}
\Gamma^\alpha_{\beta\delta}
+
B^\tau_{\ \alpha}
(\partial'_\mu A^\beta_{\ \rho})
A^\delta_{\ \nu}
\Gamma^\alpha_{\beta\delta}
$$

$$
+
B^\tau_{\ \alpha}
A^\beta_{\ \rho}
(\partial'_\mu A^\delta_{\ \nu})
\Gamma^\alpha_{\beta\delta}
+
B^\tau_{\ \alpha}
A^\beta_{\ \rho}
A^\delta_{\ \nu}
\partial'_\mu \Gamma^\alpha_{\beta\delta}.
$$

Since

$$
\partial'_\mu \Gamma^\alpha_{\beta\delta}
=
A^\gamma_{\ \mu}
\partial_\gamma \Gamma^\alpha_{\beta\delta},
$$

we obtain

$$
\partial'_\mu
\left(
B^\tau_{\ \alpha}
A^\beta_{\ \rho}
A^\delta_{\ \nu}
\Gamma^\alpha_{\beta\delta}
\right)
=
(\partial'_\mu B^\tau_{\ \alpha})
A^\beta_{\ \rho}
A^\delta_{\ \nu}
\Gamma^\alpha_{\beta\delta}
+
B^\tau_{\ \alpha}
(\partial'_\mu A^\beta_{\ \rho})
A^\delta_{\ \nu}
\Gamma^\alpha_{\beta\delta}
$$

$$
+
B^\tau_{\ \alpha}
A^\beta_{\ \rho}
(\partial'_\mu A^\delta_{\ \nu})
\Gamma^\alpha_{\beta\delta}
+
B^\tau_{\ \alpha}
A^\beta_{\ \rho}
A^\gamma_{\ \mu}
A^\delta_{\ \nu}
\partial_\gamma \Gamma^\alpha_{\beta\delta}.
$$

Similarly,

$$
\partial'_\nu
\left(
B^\tau_{\ \alpha}
A^\beta_{\ \rho}
A^\gamma_{\ \mu}
\Gamma^\alpha_{\beta\gamma}
\right)
=
(\partial'_\nu B^\tau_{\ \alpha})
A^\beta_{\ \rho}
A^\gamma_{\ \mu}
\Gamma^\alpha_{\beta\gamma}
+
B^\tau_{\ \alpha}
(\partial'_\nu A^\beta_{\ \rho})
A^\gamma_{\ \mu}
\Gamma^\alpha_{\beta\gamma}
$$

$$
+
B^\tau_{\ \alpha}
A^\beta_{\ \rho}
(\partial'_\nu A^\gamma_{\ \mu})
\Gamma^\alpha_{\beta\gamma}
+
B^\tau_{\ \alpha}
A^\beta_{\ \rho}
A^\gamma_{\ \mu}
A^\delta_{\ \nu}
\partial_\delta \Gamma^\alpha_{\beta\gamma}.
$$

Therefore,

$$
\partial'_\mu \Gamma'^\tau_{\rho\nu}
-
\partial'_\nu \Gamma'^\tau_{\rho\mu}
$$

contains the tensorial derivative part

$$
B^\tau_{\ \alpha}
A^\beta_{\ \rho}
A^\gamma_{\ \mu}
A^\delta_{\ \nu}
\left(
\partial_\gamma \Gamma^\alpha_{\beta\delta}
-
\partial_\delta \Gamma^\alpha_{\beta\gamma}
\right),
$$

plus non-tensorial terms involving derivatives of $A$ and $B$.

Now compute the quadratic Christoffel terms.

First,

$$
\Gamma'^\tau_{\lambda\mu}
=
B^\tau_{\ \alpha}
A^\eta_{\ \lambda}
A^\gamma_{\ \mu}
\Gamma^\alpha_{\eta\gamma}
+
B^\tau_{\ \alpha}
\partial'_\lambda A^\alpha_{\ \mu}.
$$

Also,

$$
\Gamma'^\lambda_{\rho\nu}
=
B^\lambda_{\ \kappa}
A^\beta_{\ \rho}
A^\delta_{\ \nu}
\Gamma^\kappa_{\beta\delta}
+
B^\lambda_{\ \kappa}
\partial'_\rho A^\kappa_{\ \nu}.
$$

Multiplying,

$$
\Gamma'^\tau_{\lambda\mu}\Gamma'^\lambda_{\rho\nu}
=
\left(
B^\tau_{\ \alpha}
A^\eta_{\ \lambda}
A^\gamma_{\ \mu}
\Gamma^\alpha_{\eta\gamma}
+
B^\tau_{\ \alpha}
\partial'_\lambda A^\alpha_{\ \mu}
\right)
\left(
B^\lambda_{\ \kappa}
A^\beta_{\ \rho}
A^\delta_{\ \nu}
\Gamma^\kappa_{\beta\delta}
+
B^\lambda_{\ \kappa}
\partial'_\rho A^\kappa_{\ \nu}
\right).
$$

Expanding,

$$
\Gamma'^\tau_{\lambda\mu}\Gamma'^\lambda_{\rho\nu}
=
B^\tau_{\ \alpha}
A^\eta_{\ \lambda}
A^\gamma_{\ \mu}
\Gamma^\alpha_{\eta\gamma}
B^\lambda_{\ \kappa}
A^\beta_{\ \rho}
A^\delta_{\ \nu}
\Gamma^\kappa_{\beta\delta}
$$

$$
+
B^\tau_{\ \alpha}
A^\eta_{\ \lambda}
A^\gamma_{\ \mu}
\Gamma^\alpha_{\eta\gamma}
B^\lambda_{\ \kappa}
\partial'_\rho A^\kappa_{\ \nu}
+
B^\tau_{\ \alpha}
\partial'_\lambda A^\alpha_{\ \mu}
B^\lambda_{\ \kappa}
A^\beta_{\ \rho}
A^\delta_{\ \nu}
\Gamma^\kappa_{\beta\delta}
$$

$$
+
B^\tau_{\ \alpha}
\partial'_\lambda A^\alpha_{\ \mu}
B^\lambda_{\ \kappa}
\partial'_\rho A^\kappa_{\ \nu}.
$$

Using

$$
A^\eta_{\ \lambda}B^\lambda_{\ \kappa}
=
\delta^\eta_{\kappa},
$$

the first term becomes

$$
B^\tau_{\ \alpha}
A^\gamma_{\ \mu}
A^\beta_{\ \rho}
A^\delta_{\ \nu}
\Gamma^\alpha_{\kappa\gamma}
\Gamma^\kappa_{\beta\delta}.
$$

Renaming dummy indices,

$$
B^\tau_{\ \alpha}
A^\beta_{\ \rho}
A^\gamma_{\ \mu}
A^\delta_{\ \nu}
\Gamma^\alpha_{\lambda\gamma}
\Gamma^\lambda_{\beta\delta}.
$$

Similarly,

$$
\Gamma'^\tau_{\lambda\nu}\Gamma'^\lambda_{\rho\mu}
$$

has tensorial quadratic part

$$
B^\tau_{\ \alpha}
A^\beta_{\ \rho}
A^\gamma_{\ \mu}
A^\delta_{\ \nu}
\Gamma^\alpha_{\lambda\delta}
\Gamma^\lambda_{\beta\gamma}.
$$

Therefore,

$$
\Gamma'^\tau_{\lambda\mu}\Gamma'^\lambda_{\rho\nu}
-
\Gamma'^\tau_{\lambda\nu}\Gamma'^\lambda_{\rho\mu}
$$

contains the tensorial quadratic part

$$
B^\tau_{\ \alpha}
A^\beta_{\ \rho}
A^\gamma_{\ \mu}
A^\delta_{\ \nu}
\left(
\Gamma^\alpha_{\lambda\gamma}
\Gamma^\lambda_{\beta\delta}
-
\Gamma^\alpha_{\lambda\delta}
\Gamma^\lambda_{\beta\gamma}
\right).
$$

Combining this with the tensorial derivative part, we obtain

$$
B^\tau_{\ \alpha}
A^\beta_{\ \rho}
A^\gamma_{\ \mu}
A^\delta_{\ \nu}
\left(
\partial_\gamma \Gamma^\alpha_{\beta\delta}
-
\partial_\delta \Gamma^\alpha_{\beta\gamma}
+
\Gamma^\alpha_{\lambda\gamma}
\Gamma^\lambda_{\beta\delta}
-
\Gamma^\alpha_{\lambda\delta}
\Gamma^\lambda_{\beta\gamma}
\right).
$$

But the expression in parentheses is precisely

$$
R^\alpha_{\ \beta\gamma\delta}.
$$

Hence the tensorial part is

$$
B^\tau_{\ \alpha}
A^\beta_{\ \rho}
A^\gamma_{\ \mu}
A^\delta_{\ \nu}
R^\alpha_{\ \beta\gamma\delta}.
$$

It remains to see why the remaining terms cancel.

The remaining terms contain derivatives of the coordinate transformation matrices $A$ and $B$. They cancel because $A$ and $B$ are inverse Jacobian matrices. Since

$$
B^\tau_{\ \alpha}A^\alpha_{\ \rho}
=
\delta^\tau_{\rho},
$$

differentiating with respect to $x'^\mu$ gives

$$
(\partial'_\mu B^\tau_{\ \alpha})A^\alpha_{\ \rho}
+
B^\tau_{\ \alpha}(\partial'_\mu A^\alpha_{\ \rho})
=
0.
$$

Therefore,

$$
(\partial'_\mu B^\tau_{\ \alpha})A^\alpha_{\ \rho}
=
-
B^\tau_{\ \alpha}(\partial'_\mu A^\alpha_{\ \rho}).
$$

This identity is exactly what cancels the terms produced by differentiating the Jacobian factors against the mixed products coming from

$$
\Gamma'^\tau_{\lambda\mu}\Gamma'^\lambda_{\rho\nu}
-
\Gamma'^\tau_{\lambda\nu}\Gamma'^\lambda_{\rho\mu}.
$$

Finally, the purely second-derivative coordinate terms cancel because

$$
\partial'_\mu\partial'_\rho A^\tau_{\ \nu}
-
\partial'_\nu\partial'_\rho A^\tau_{\ \mu}
=
0,
$$

since mixed partial derivatives commute:

$$
\partial'_\mu\partial'_\rho A^\tau_{\ \nu}
=
\partial'_\rho\partial'_\mu A^\tau_{\ \nu},
\qquad
\partial'_\nu\partial'_\rho A^\tau_{\ \mu}
=
\partial'_\rho\partial'_\nu A^\tau_{\ \mu}.
$$

Thus all non-tensorial contributions cancel, and only the tensorial expression remains:

$$
R'^\tau_{\ \rho\mu\nu}
=
B^\tau_{\ \alpha}
A^\beta_{\ \rho}
A^\gamma_{\ \mu}
A^\delta_{\ \nu}
R^\alpha_{\ \beta\gamma\delta}.
$$

Restoring the definitions of $A$ and $B$,

$$
R'^\tau_{\ \rho\mu\nu}
=
\frac{\partial x'^\tau}{\partial x^\alpha}
\frac{\partial x^\beta}{\partial x'^\rho}
\frac{\partial x^\gamma}{\partial x'^\mu}
\frac{\partial x^\delta}{\partial x'^\nu}
R^\alpha_{\ \beta\gamma\delta}.
$$

Therefore the Riemann curvature object transforms as a genuine tensor of type $(1,3)$.
